```python
sales = customers_orders.join(products,on="ProductID",how="inner")
```

```python
sales.printSchema()
```

```python
sales.show(5)
```

```python
#1. Display customer names along with the products they purchased.
sales.select("CustomerName","ProductName").show()
```

```python
#2. Find the total number of orders placed by each customer.
sales.groupby("CustomerID").count().show()
```

```python
#3. Find the total quantity purchased by each customer.
sales.groupby("CustomerID").agg(sum("Quantity")).show()
```

```python
#4. Find the top 10 customers who purchased the highest quantity of products.
sales.groupby("CustomerID").agg(sum("Quantity").alias("sumOfQuantity")).orderBy(desc("sumOfQuantity")).show(10)
```

```python
#5. Find the total quantity sold for each product.
sales.groupby("ProductID").agg(sum("Quantity")).show()
```

```python
#6. Find the most expensive product in each category.
sales.groupby("category").agg(max("price")).show()
```

```python
#7. Find the number of customers in each city.
sales.groupby("city").agg(count("customerId")).show()
```

```python
#8. Calculate the total sales amount for each customer.
sales.groupby("customerId").agg(sum(col("price")*col("Quantity")).alias("Total")).show()
```

```python
#9 Which city generated the highest sales.
sales.groupby("city").agg(sum(col("price")*col("Quantity")).alias("Total")).orderBy(desc("Total")).show(1)
```

```python
#10 Which product was ordered the most times.
sales.groupby("ProductName").agg(sum("Quantity").alias("max")).orderBy(desc("max")).show(1)
```

```python
#11 Find customers who purchased products from more than one category.
sales.groupBy("CustomerName") \
     .agg(countDistinct("Category").alias("NumberOfCategories")) \
     .filter(col("NumberOfCategories") > 1) \
     .orderBy(col("NumberOfCategories").desc()) \
     .show()
```

```python
#12 Top 5 products that generated the highest revenue
sales.groupby("ProductName").agg(sum(col("price") * col("Quantity")).alias("Total")).orderBy(desc("Total")).show(5)
```
